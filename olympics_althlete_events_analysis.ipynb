```python
# Import needed packages
import numpy as np 
import pandas as pd 

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('../input/olympics-althlete-events-analysis/athlete_events.csv')
df.head()
```

```python
df.describe()
```

```python
# Eliminate rows that have missing values
df.dropna(axis=0, inplace= False)
```

```python
df.hist(figsize=(20,10))
```

Let's see the numbers of medals

```python
gold = df.groupby(['Medal'])['ID'].count()
gold
```

Numbers of players in each game

```python
nbplayer = df.groupby(['Sport'])['ID'].count()
nbplayer
```

```python
import matplotlib.pyplot as plt
nbplayer = df.groupby(['Sport'])['ID'].count().plot(kind='bar', figsize=(20,10),ax=plt.gca())
```

```python
# comparison of men and women
sex = df.groupby(['Sex'])['ID'].count().plot(kind='pie', figsize=(8,8),ax=plt.gca())
```

```python
tabmedal = pd.pivot_table(df, index=['Name'] ,columns=['Medal'], aggfunc={'Medal': 'size'}, fill_value=0)
tabmedal
```

Let's see the distribution of the age of players 

```python
sex = df.groupby(['Age'])['ID'].count().plot(kind='line', figsize=(15,10),ax=plt.gca())
```

Let's see the distribution of players in each city

```python
nbplayer = df.groupby(['City'])['ID'].count().plot(kind='bar', figsize=(20,10),ax=plt.gca())
```
